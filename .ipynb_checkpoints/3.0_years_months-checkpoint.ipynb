```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('attacks2.csv')
```

# Cleaning data

```python
df['Year'].replace({3019:2019},inplace=True)

df['Fatal (Y/N)']=df['Fatal (Y/N)'].fillna('No data')

df.loc[df['Fatal (Y/N)']=='y','Fatal (Y/N)']= 'Y'

for i in df['Fatal (Y/N)']:
    if i in ['F','M','2017','UNKNOWN']:
        df.loc[df['Fatal (Y/N)']==i,'Fatal (Y/N)']= 'No data'
        
df.loc[df['Fatal (Y/N)']=='Y','Fatal (Y/N)']= 'Fatal'
df.loc[df['Fatal (Y/N)']=='N','Fatal (Y/N)']= 'Non fatal'
```

```python
df['Fatal (Y/N)'].value_counts()
```

```python
len(df[df['Year']==0])
```

```python
df['Year'].replace({0:None},inplace=True)
```

# Years

```python
sns.countplot(x=df['Year'][df['Year']>1850])
```

```python
sns.lineplot(x=df['Year'].value_counts().index,
             y=df['Year'].value_counts().values, 
             linewidth=2.5)
```

```python
dfy = pd.DataFrame(df['Year'].value_counts())
dfy.reset_index(inplace=True)

dfy.rename(columns={'index':'Year','Year':'Counts'},inplace=True)
```

```python
sns.lineplot(x='Year',
             y='Counts',
             data=dfy,
             linewidth=0.5).set(title='Shark attacks through the years')
```

```python
#sns.set(rc={'figure.figsize':(11.7,8.27)})

sns.lineplot(x='Year',y='Counts',data=dfy,linewidth=1).set(title='Shark attacks since 1800')
#plt.savefig('years.png')
```

# Fatal / Non Fatal

```python
bar = sns.barplot(data=df, 
                  x=df['Fatal (Y/N)'].value_counts().index, 
                  y=df['Fatal (Y/N)'].value_counts().values,
                  palette='Blues_r')


for container in bar.containers:
    bar.bar_label(container)

bar.set(title='Total fatal/Non fatal attacks')
    
#plt.savefig('fatal.png')
```

```python
df = df[df['Year']>=1800]
```

```python
df['Decade'] = 10 * (df['Year'] // 10)
```

```python
df['Decade'] = df['Decade'].astype('int')
```

```python
ax = sns.countplot(data=df,
                   x=df['Decade'].sort_values(),
                   palette='Blues',
                   hue=df['Fatal (Y/N)'])

ax2 = ax.set_xticklabels(ax.get_xticklabels(),rotation=90)

ax.set(title='Fatal/Non fatal shark attacks for each decade since 1800')

#plt.savefig('fatals_decade.png')
```

# Provoked / Unprovoked

```python
df = df.replace({'Type' : {'Questionable':'Unverified', 'Unconfirmed':'Unverified', 'Invalid':'Unverified','Under investigation':'Unverified',
                          'Sea Disaster':'Boating','Boat':'Boating','Watercraft':'Boating','Boatomg':'Boating'}})
```

```python
ax = sns.countplot(data=df,
                   x=df['Type'],
                   palette='Blues_r')

for container in ax.containers:
    ax.bar_label(container)
    
ax.set(title='Provoked/Unprovoked shark attacks for all reports')
#plt.savefig('prov_total.png')
```

```python
ax = sns.countplot(data=df,
                   x=df['Decade'].sort_values(),
                   palette='Blues_r',
                   hue=df['Type'])

ax2 = ax.set_xticklabels(ax.get_xticklabels(),rotation=90)

ax.set(title='Provoked/Unprovoked shark attacks for each decade since 1800')

#plt.savefig('provoked_decade.png')
```

# Months

```python
df['Date']=df['Date'].str.replace('Reported ','')
```

```python
df['Month']= df["Date"].str[3:6]
```

```python
months_list = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Unknown']
```

```python
def clear_months(month):
    if month not in months_list:
        return 'Unknown'
    else:
        return month
```

```python
df['Month']=df['Month'].apply(clear_months)
```

```python
mon = sns.countplot(data=df,
                    x=df['Month'],
                    order=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Unknown'],
                    palette='Blues')


for container in mon.containers:
    mon.bar_label(container)

mon.set(title='Shark attacks for each month')
    
mon2 = mon.set_xticklabels(mon.get_xticklabels(),rotation=20)
    
fig = mon.get_figure()
#fig.savefig("months.png")
```

```python
mon = sns.countplot(data=df,
                    x=df['Month'],
                    order=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Unknown'],
                    palette='Blues',
                    hue=df['Fatal (Y/N)'])

mon.set(title='Fatal/Non fatal attacks for each month')
    
mon2 = mon.set_xticklabels(mon.get_xticklabels(),rotation=45)
    
fig = mon.get_figure()
#fig.savefig("months_fatal.png",dpi=300, bbox_inches='tight')
```
